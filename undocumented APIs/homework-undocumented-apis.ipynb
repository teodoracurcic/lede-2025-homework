```python
responses = [] 
bedrooms = [0, 1, 2, 3, 4, 5, 6, 7, 8]

for address in addresses:
    for bedroom in bedrooms:

        params = {
        'operationName': 'GetHostEstimateData',
        'locale': 'en',
        'variables': '{"durationGranularity":["MONTHLY"],"source":"HOST_LANDING_PAGE","location":{"searchQuery":"' + address + '"},"roomTypeCategory":"ENTIRE_HOME", "bedroom":' + str(bedroom) + ',"fetchDebugInfo":false}',
        'extensions': '{"persistedQuery":{"version":1,"sha256Hash":"7aa6bb483a31d77a137c3442b007c846a3c0ab778c2581ace8f264c0fe10648b"}}',
        }

        response = requests.get(url, params=params, headers=headers)
        
        # check if the API call was successful by checking the status
        if response.status_code == 200:
            data = response.json()
            data['bedroom'] = bedroom
            data['address'] = address
            responses.append(data)
    
        # put some time between requests
        time.sleep(1) 
```

```python
for record in responses:
    try:
        screen = record['data']['presentation']['hostEstimate']['hostEstimateScreen']
        nightly = screen['header']['sections'][1]['value']
        address = record['address']
        bedroom = record['bedroom']
        print(f"{address}, {bedroom} bedrooms: {nightly} / night")
    except Exception as e:
        print(f"Skipped: {e}")
```

```python
#motnhly earnings from airbnb

for record in responses:
    try:
        screen = record['data']['presentation']['hostEstimate']['hostEstimateScreen']
        monthly = screen['header']['sections'][0]['value']   
        address = record['address']
        bedroom = record['bedroom']
        print(f"{address}, {bedroom} bedrooms:, {monthly} / month")
    except Exception as e:
        print(f"⚠️ Skipped: {e}")
```

```python
#converting to df

results = []

for record in responses:
    try:
        screen = record['data']['presentation']['hostEstimate']['hostEstimateScreen']
        nightly = screen['header']['sections'][1]['value']
        monthly = screen['header']['sections'][0]['value']
        address = record['address']
        bedroom = record['bedroom']
        results.append({
            'address': address,
            'bedrooms': bedroom,
            'per_month': monthly,
            'per_night': nightly
        })
    except:
        pass
```

```python
import pandas as pd
df = pd.DataFrame(results)
```
